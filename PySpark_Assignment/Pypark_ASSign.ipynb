```python
employee_data = [
    (101, "Alice", 25, "HR", 45000, "Chennai", "2022-01-15", 201),
    (102, "Bob", 30, "IT", 70000, "Bangalore", "2021-06-20", 202),
    (103, "Charlie", None, "IT", None, "Chennai", "2023-03-12", 202),
    (104, "David", 28, "Finance", 65000, "Mumbai", "2020-09-18", 203),
    (105, "Eva", 35, "HR", 80000, None, "2019-05-25", 201),
    (106, "Frank", 29, "Marketing", 55000, "Hyderabad", None, 204),
    (107, "Grace", 31, "Finance", None, "Pune", "2022-12-01", 203),
    (108, "Henry", 26, "IT", 60000, "Bangalore", "2024-01-10", None)
]

employee_columns = [
    "emp_id",
    "name",
    "age",
    "department",
    "salary",
    "city",
    "joining_date",
    "manager_id"
]

employee_df = spark.createDataFrame(employee_data, employee_columns)

employee_df.show()
employee_df.printSchema()
display(employee_df)
```

```python
department_data = [
    ("HR", "Chennai"),
    ("IT", "Bangalore"),
    ("Finance", "Mumbai"),
    ("Marketing", "Hyderabad"),
    ("Sales", "Delhi")
]

department_columns = [
    "dept_name",
    "location"
]

department_df = spark.createDataFrame(department_data, department_columns)

department_df.show()
department_df.printSchema()
display(department_df)
```

```python
manager_data = [
    (201, "Robert"),
    (202, "Jennifer"),
    (203, "Michael"),
    (204, "Sophia")
]

manager_columns = [
    "manager_id",
    "manager_name"
]

manager_df = spark.createDataFrame(manager_data, manager_columns)

manager_df.show()
manager_df.printSchema()
display(manager_df)
```

```python
from pyspark.sql.functions import *

from pyspark.sql.window import Window
```

```python
# 4. Display the schema of all DataFrames
employee_df.printSchema()
department_df.printSchema()
manager_df.printSchema()

# 5. Display the first 5 rows
employee_df.show(5)
department_df.show(5)
manager_df.show(5)
```

```python
#Paert B
# 6. Select only emp_id, name, and salary
employee_df.select("emp_id", "name", "salary").show()

# 7. Select all employees from the IT department
employee_df.filter(col("department") == "IT").show()

# 8. Select employee names and alias the column as Employee_Name
employee_df.select(col("name").alias("Employee_Name")).show()

# 9. Select salary after increasing it by 10%
employee_df.select("name", "salary", (col("salary") * 1.10).alias("Increased_salary")).show()

# 10. Select employees whose city is Chennai or Bangalore
employee_df.filter(col("city").isin("Chennai", "Bangalore")).show()
```

```python
#Part C

# 11. Create a new column called bonus equal to 20% of salary
employee_df.withColumn("bonus", col("salary") * 0.20).show()

# 12. Create a column called annual_salary
employee_df.withColumn("annual_salary", col("salary") * 12).show()

# 13. Create a column called experience_bonus
employee_df.withColumn("experience_bonus", 
    when(col("salary") > 70000, 10000)
    .when(col("salary").between(50000, 70000), 5000)
    .otherwise(2000)
).show()

# 14. Convert employee names to uppercase
employee_df.withColumn("name", upper(col("name"))).show()

# 15. Add a column called country with value "India"
employee_df.withColumn("country", lit("India")).show()
```

```python
# 16. Rename department to dept
employee_df.withColumnRenamed("department", "dept").show()

# 17. Rename salary to monthly_salary
employee_df.withColumnRenamed("salary", "monthly_salary").show()

# 18. Rename multiple columns one by one
employee_df.withColumnRenamed("name", "employee_name") \
           .withColumnRenamed("department", "dept") \
           .withColumnRenamed("salary", "monthly_salary") \
           .withColumnRenamed("city", "location").show()
```

```python
# 19. Find employees earning more than 60000
employee_df.filter(col("salary") > 60000).show()

# 20. Find employees older than 30
employee_df.filter(col("age") > 30).show()

# 21. Find HR employees in Chennai
employee_df.filter((col("department") == "HR") & (col("city") == "Chennai")).show()

# 22. Find employees with NULL salary
employee_df.filter(col("salary").isNull()).show()

# 23. Find employees whose salary is between 50000 and 70000
employee_df.filter(col("salary").between(50000, 70000)).show()

# 24. Find employees whose names start with "A"
employee_df.filter(col("name").startswith("A")).show()

# 25. Find employees whose city is not NULL
employee_df.filter(col("city").isNotNull()).show()
```

```python
# 26. Sort employees by salary ascending
employee_df.orderBy(col("salary").asc()).show()

# 27. Sort employees by salary descending
employee_df.orderBy(col("salary").desc()).show()

# 28. Sort employees by department and salary
employee_df.orderBy(col("department").asc(), col("salary").desc()).show()

# 29. Sort employees by age descending
employee_df.orderBy(col("age").desc()).show()

# 30. Display top 3 highest-paid employees
employee_df.orderBy(col("salary").desc()).limit(3).show()
```

```python
# Part G
# 31. Replace NULL salary with 30000
employee_df.na.fill(30000, subset=["salary"]).show()

# 32. Replace NULL city with "Unknown"
employee_df.na.fill("Unknown", subset=["city"]).show()

# 33. Replace NULL age with average age
avg_age = employee_df.select(avg("age")).collect()[0][0]
employee_df.na.fill(int(avg_age), subset=["age"]).show()

# 34. Replace multiple NULL columns in one statement
employee_df.na.fill({"salary": 30000, "city": "Unknown","age":employee_df.select(avg("age")).collect()[0][0]}).show()

# 35. Drop rows containing any NULL values
employee_df.na.drop(how="any").show()

# 36. Drop rows where salary is NULL
employee_df.na.drop(subset=["salary"]).show()

# 37. Drop rows only if all values are NULL
employee_df.na.drop(how="all").show()
```

```python
# Part H
# 38. Convert joining_date to DateType
employee_df = employee_df.withColumn("joining_date", to_date(col("joining_date"), "yyyy-MM-dd"))

# 39. Extract joining year
employee_df.select("name", year("joining_date").alias("joining_year")).show()

# 40. Extract joining month
employee_df.select("name", month("joining_date").alias("joining_month")).show()

# 41. Extract joining day
employee_df.select("name", dayofmonth("joining_date").alias("joining_day")).show()

# 42. Find employees who joined after 2022
employee_df.filter(year("joining_date") > 2022).show()

# 43. Calculate years worked
employee_df.withColumn("years_worked", round(months_between(current_date(), col("joining_date")) / 12, 1)).show()

# 44. Find employees who joined this year
employee_df.filter(year("joining_date") == year(current_date())).show()

# 45. Display current date
employee_df.select(current_date().alias("current_date")).limit(1).show()

# 46. Display current timestamp
employee_df.select(current_timestamp().alias("current_timestamp")).limit(1).show()

# 47. Add 30 days to joining date
employee_df.select("name", date_add("joining_date", 30).alias("date_plus_30")).show()

# 48. Find difference between joining date and today's date
employee_df.select("name", datediff(current_date(), "joining_date").alias("days_worked")).show()
```

```python
# Part I
#  49. Inner join between Employee and Department
employee_df.join(department_df, employee_df.department == department_df.dept_name, "inner").show()

# 50. Left join
employee_df.join(department_df, employee_df.department == department_df.dept_name, "left").show()

# 51. Right join
employee_df.join(department_df, employee_df.department == department_df.dept_name, "right").show()

# 52. Full outer join
employee_df.join(department_df, employee_df.department == department_df.dept_name, "outer").show()

# 53. Join Employee with Manager DataFrame
employee_df.join(manager_df, "manager_id", "left").show()

# 54. Display employee name along with manager name
employee_df.join(manager_df, "manager_id", "left").select("name", "manager_name").show()

# 55. Find employees whose department location matches their city
employee_df.join(department_df, employee_df.department == department_df.dept_name) \
           .filter(col("city") == col("location")).show()

# 56. Find departments having no employees
department_df.join(employee_df, department_df.dept_name == employee_df.department, "left_anti").show()
```

```python
# Part J
# Create additional DataFrame for Union
new_employee_data = [
    (109, "Irene", 27, "HR", 52000, "Chennai", "2024-03-11", 201),
    (110, "Jack", 34, "Sales", 72000, "Delhi", "2021-11-20", 205)
]
new_employee_df = spark.createDataFrame(new_employee_data, employee_columns)

# 57. Union the two DataFrames
combined_df = employee_df.union(new_employee_df)
combined_df.show()

# 58. Count total employees after union
print("Total Count:", combined_df.count())

# 59. Remove duplicate employees after union
combined_df.dropDuplicates().show()

# 60. Union by column names
employee_df.unionByName(new_employee_df).show()
```

```python
# Part K
# 61. Display distinct departments
employee_df.select("department").distinct().show()

# 62. Count employees in each department
employee_df.groupBy("department").count().show()

# 63. Calculate average salary by department
employee_df.groupBy("department").avg("salary").show()

# 64. Find maximum salary by department
employee_df.groupBy("department").max("salary").show()

# 65. Find minimum salary by department
employee_df.groupBy("department").min("salary").show()

# 66. Find total salary paid in each department
employee_df.groupBy("department").sum("salary").show()

# 67. Count employees in each city
employee_df.groupBy("city").count().show()

# 68. Display unique cities
employee_df.select("city").distinct().show()

# 69. Remove duplicate records
employee_df.dropDuplicates().show()

# 70. Drop the manager_id column
employee_df.drop("manager_id").show()
```

```python
# Part L
# 71. Convert employee names to lowercase
employee_df.select(lower(col("name")).alias("lowercase_name")).show()

# 72. Find length of each employee name
employee_df.select("name", length(col("name")).alias("name_length")).show()

# 73. Extract first three letters of employee names
employee_df.select("name", substring(col("name"), 1, 3).alias("short_name")).show()

# 74. Concatenate employee name and city
employee_df.select(concat_ws(" - ", col("name"), col("city")).alias("name_city")).show()

# 75. Replace all occurrences of "a" with "@" in names
employee_df.select("name", regexp_replace(col("name"), "a", "@").alias("modified_name")).show()
```

```python
# Part M
# 76. Find total salary
employee_df.select(sum("salary").alias("total_salary")).show()

# 77. Find average salary
employee_df.select(avg("salary").alias("average_salary")).show()

# 78. Find highest-paid employee
employee_df.orderBy(col("salary").desc()).limit(1).show()

# 79. Find lowest-paid employee
employee_df.filter(col("salary").isNotNull()).orderBy(col("salary").asc()).limit(1).show()

# 80. Count employees with salary greater than 50000
print("Count:", employee_df.filter(col("salary") > 50000).count())
```

```python
# Part N
# 81. Display employees whose salary is above department average
w_dept = Window.partitionBy("department")
employee_df.withColumn("dept_avg", avg("salary").over(w_dept)) \
           .filter(col("salary") > col("dept_avg")).drop("dept_avg").show()

# 82. Find second highest salary
w_all = Window.orderBy(col("salary").desc())
employee_df.filter(col("salary").isNotNull()) \
           .withColumn("rnk", dense_rank().over(w_all)) \
           .filter(col("rnk") == 2).drop("rnk").show()

# 83. Find duplicate cities
employee_df.groupBy("city").count().filter(col("count") > 1).show()

# 84. Display employees who joined in the last two years
employee_df.filter(year("joining_date") >= (year(current_date()) - 2)).show()

# 85. Replace NULL manager IDs with 999
employee_df.na.fill(999, subset=["manager_id"]).show()

# 86. Create a salary grade
employee_df.withColumn("salary_grade", 
    when(col("salary") >= 80000, "A")
    .when(col("salary") >= 60000, "B")
    .when(col("salary") >= 40000, "C")
    .otherwise("D")
).show()

# 87. Create an employee ID string like EMP101
employee_df.withColumn("emp_str_id", concat(lit("EMP"), col("emp_id"))).show()

# 88. Find employees whose names end with "e"
employee_df.filter(col("name").endswith("e")).show()

# 89. Display employees sorted by joining date
employee_df.orderBy("joining_date").show()

# 90. Find employees who have worked for more than three years
employee_df.filter((months_between(current_date(), col("joining_date")) / 12) > 3).show()
```

```python
# 91. Rank employees by salary within each department
w_dept_sal = Window.partitionBy("department").orderBy(col("salary").desc())
employee_df.withColumn("rank", rank().over(w_dept_sal)).show()

# 92. Find top 2 highest-paid employees in every department
employee_df.withColumn("rank", dense_rank().over(w_dept_sal)).filter(col("rank") <= 2).show()

# 93. Find departments with average salary greater than 60000
employee_df.groupBy("department").avg("salary").filter(col("avg(salary)") > 60000).show()

# 94. Find employees who do not have a manager
employee_df.filter(col("manager_id").isNull()).show()

# 95. Display employees with manager names and department locations in a single DataFrame
employee_df.join(manager_df, "manager_id", "left") \
           .join(department_df, employee_df.department == department_df.dept_name, "left") \
           .select("name", "manager_name", "location").show()

# 96. Identify employees with missing information (NULL values)
employee_df.filter(isnull("salary") | isnull("age") | isnull("city") | isnull("joining_date") | isnull("manager_id")).show()

# 97. Replace missing salaries with department average salary
employee_df.withColumn("dept_avg", avg("salary").over(w_dept)) \
           .withColumn("salary", when(col("salary").isNull(), col("dept_avg")).otherwise(col("salary"))) \
           .drop("dept_avg").show()

# 98. Calculate monthly tax as 10% of salary
employee_df.withColumn("monthly_tax", col("salary") * 0.10).show()

# 99. Find employees who joined on weekends (1=Sunday, 7=Saturday)
employee_df.filter(dayofweek("joining_date").isin(1, 7)).show()

# 100. Create a final report
final_report = employee_df.join(manager_df, "manager_id", "left") \
    .join(department_df, employee_df.department == department_df.dept_name, "left") \
    .withColumn("bonus", col("salary") * 0.20) \
    .withColumn("annual_salary", col("salary") * 12) \
    .withColumn("years_worked", round(months_between(current_date(), col("joining_date")) / 12, 1)) \
    .withColumn("salary_grade", 
        when(col("salary") >= 80000, "A")
        .when(col("salary") >= 60000, "B")
        .when(col("salary") >= 40000, "C")
        .otherwise("D")) \
    .select(
        col("name").alias("Employee Name"),
        col("department").alias("Department"),
        col("manager_name").alias("Manager Name"),
        col("location").alias("Department Location"),
        col("salary").alias("Salary"),
        col("bonus").alias("Bonus"),
        col("annual_salary").alias("Annual Salary"),
        col("years_worked").alias("Years Worked"),
        col("salary_grade").alias("Salary Grade")
    )

final_report.show()
```
